### Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [0, 5]
y = [0, 200]

plt.plot(x,y)
plt.show()
```

```python
x = [0, 5]
y = [0, 200]

plt.plot(x, y, 'x')

plt.show()
```

```python
x = [0, 1, 3, 5]
y = [0, 10, 85, -30]

plt.plot(x, y,)

plt.show()
```

```python
points = [0, 10, 85, 7, 110, -30]

plt.plot(points, '--', color='g')#it considers single array input as Y axis and its index as X axis

plt.show()
```

```python
points = [0, 10, 85, 7, 110, -30]

plt.plot(points, '--', color='g')
#Graph labels
plt.xlabel("Xaxis")
plt.ylabel("Yaxis")
plt.title('Graph')
plt.show()
```

```python
font1 = {'family':'serif', 'color':'red', 'size':20}
font2 = {'family':'serif', 'color':'blue', 'size':10}
points = [0, 10, 85, 7, 110, -30]

plt.plot(points, '--', color='g')
#Graph labels
plt.xlabel("Xaxis", fontdict=font2)
plt.ylabel("Yaxis", fontdict=font1)
plt.title('Graph')
plt.show()
```

```python
import pandas as pd
```

```python
bikes = pd.read_csv('Bike_Sharing_day.csv')
```

```python
bikes.head()
```

```python
dte = pd.to_datetime(bikes.dteday)
```

```python
bikes['year'] = dte.dt.year
```

```python
bikes.head()
```

```python
plt.plot(bikes.season,'x')
```

```python
plt.plot(bikes.temp,'x')
```

```python
plt.plot(bikes.cnt,'x')
```

```python
plt.plot(bikes.season, bikes.cnt, 'o')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

```python
plt.plot(bikes.holiday, bikes.cnt, 'o')
plt.xlabel('Holiday')
plt.ylabel('Count')
plt.show()
```

```python
season_group = bikes.groupby('season')
count = season_group.cnt.sum()
plt.plot(bikes.season.unique(), bikes.groupby('season').cnt.sum(), 'o')
```

```python
plt.ticklabel_format(style='plain')#avoid implicit conversion
plt.plot(bikes.season.unique(), bikes.groupby('season').cnt.sum(), 'o')
```

```python
plt.ticklabel_format(style='plain')
plt.plot(bikes.year.unique(), bikes.groupby('year').cnt.sum() , 'x')
```

Other Visualisation

```python
fig, ax = plt.subplots()

index = bikes.season.unique()
value = bikes.groupby('season')['cnt'].sum()

bar_width = 0.2
ax.bar(index, value, bar_width)

ax.bar(index, bikes.groupby('season').casual.sum(), bar_width)
```

```python
fig, ax = plt.subplots()

ax.scatter(bikes.casual, bikes.registered)
```

```python
fig, ax = plt.subplots()

values = bikes.groupby('season')['cnt'].sum()
labels = ['s1', 's2', 's3', 's4']

ax.pie(values, labels=labels)
plt.show()
```

```python
sns.scatterplot(data=tips, x=tips['sex'],  y=tips['tip'])
```

```python
sns.scatterplot(data=tips, x='sex',  y='tip')
```

```python
sns.scatterplot(data=tips, x = 'total_bill', y = 'tip')
```

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip')
```

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', kind='line')
```

```python
sns.regplot(data=tips, x = 'total_bill', y = 'tip')
```

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', hue='time')
```

```python
help(sns.relplot)
```

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', hue = 'time', style='smoker')
```

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', row='day')
```

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', col='day', hue='time')
```

```python
male_data = tips[tips['sex'] == "Male"]
sns.relplot(data=male_data , x = 'total_bill', y = 'tip', hue='time')
```

```python
female_data = tips[tips['sex'] == "Female"]
sns.relplot(data=female_data , x = 'total_bill', y = 'tip', hue='time')
```
