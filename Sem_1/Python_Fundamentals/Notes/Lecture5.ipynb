Add this year series to the bikes dataset

group casual riders by seasons

```python
bikes_seasons = bikes.groupby('season')
```

```python
season_casual_sum = bikes_seasons.casual.sum()
season_rgistered_sum = bikes_seasons.registered.sum()
```

concat

```python
frames = [season_casual_sum, season_rgistered_sum]
```

```python
df = pd.concat(frames, axis = 1)
```

```python
df['total'] = season_casual_sum + season_rgistered_sum
```
