```python
font1 = { 'family' : 'serif' , 'color' : 'red', 'size' : 20}
font2 = { 'family' : 'serif' , 'color' : 'blue', 'size' : 10}

points = np.array([3,5,2,11,4,8])
plt.plot(points, linestyle='dashed', color='g')
plt.xlabel("This is the x-axis", fontdict=font2)
plt.ylabel("This is the y-axis", fontdict=font2)
plt.title("This is my first plot ...", fontdict=font1)
```

<b> Bivariate analysis <b> 

```python
bikes.columns
```

```python
plt.plot(bikes.season,bikes.cnt, 'o')
```

```python
bikes.holiday.unique()
```

```python
plt.plot(bikes.holiday, bikes.cnt, 'o')
```

```python
df_by_seasons = bikes.groupby("season")
```

```python
df_seasons = df_by_seasons.cnt.sum().sort_values()
```

```python
df_seasons
```

```python
plt.plot(bikes.season.unique(), df_by_seasons.cnt.sum(), 'o')
```

```python
plt.ticklabel_format(style='plain')
plt.plot(bikes.season.unique(), df_by_seasons.cnt.sum(), 'o')
```

```python
df_by_year = bikes.groupby("year")
```

```python
plt.ticklabel_format(style='plain')
plt.plot(bikes.year.unique(), df_by_year.cnt.sum(), 'o' )
```

<b> Other visualizations <b>

```python
fig, ax = plt.subplots()

index = bikes.season.unique()
values = df_by_seasons.cnt.sum()

bar_width= 0.25

ax.bar(index,values, bar_width)
```

```python
fig , ax = plt.subplots()

ax.scatter(bikes.casual, bikes.registered)
plt.show
```

```python
fig, ax = plt.subplots()

values = df_by_seasons.cnt.sum()
labelvalues = ["s1", "s2", "s3", "s4"]

ax.pie(values, labels=labelvalues)
plt.show()
```
