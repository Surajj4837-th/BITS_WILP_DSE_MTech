```python
import pandas as pd 
import seaborn as sns
```

```python
sns.__version__
```

```python
tips = sns.load_dataset("tips")
```

```python
type(tips)
```

```python
tips.columns
```

```python
tips.head(3)
```

```python
tips.info()
```

```python
sns.scatterplot(data = tips, x= tips["sex"], y = tips["tip"])
```

```python
sns.scatterplot(data = tips, x= "sex", y = "tip")
```

```python
sns.scatterplot(data = tips , x = "total_bill", y="tip")
```

```python
sns.lineplot(data=tips, x="total_bill", y="tip")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind="line")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip" , hue="time")
```

```python
help(sns.relplot)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip" , hue="time", style="smoker")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip" , col="day")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip" , col="day", hue="time")
```

```python
male = tips[tips.sex == "Male"]
```

```python
male
```

```python
sns.relplot(data=male, x= "total_bill" , y="tip", hue="time" )
```

```python
female = tips[tips.sex == "Female"]
```

```python
sns.relplot(data=female, x= "total_bill" , y="tip", hue="time" )
```
